# Effects of various K values:

In KNN regression, the predicted value at any point is simply the average of the K nearest neighbors’ values. The value of K determines how "smooth" or "wiggly" the prediction curve will be.

As the value of K increases in KNN regression, the prediction curve becomes progressively smoother. With K=3, the model closely follows the local variations in the data, producing a more flexible and responsive curve. At K=5, the curve still captures the overall trend while reducing sensitivity to noise, offering a balanced representation of the data. However, with K=7 and especially K=9, the curve becomes overly smoothed, losing important details and failing to capture smaller fluctuations in the data. This results in a more generalized and flatter curve that may not reflect the true structure of the underlying data.

# KNN-Classification-Model

The K-nearest neighbors (KNN) classification model was implemented by first computing the Euclidean distances between each test point and all training samples, then sorting these distances to identify the k closest neighbors. Unlike KNN regression, which averages target values, the classification variant predicts the output class by majority voting—selecting the most frequent label among the nearest k points. The model’s performance was evaluated across different k values (3, 5, 7, and 9), revealing a clear trade-off between flexibility and stability. Smaller k values (e.g., 3) produced more complex decision boundaries that closely fit the training data but were sensitive to noise, while larger k values (e.g., 9) led to smoother, more generalized boundaries at the cost of potentially oversimplifying local patterns. Intermediate values like 5 or 7 often struck a balance, mitigating overfitting without sacrificing too much detail. The results underscored the importance of selecting an appropriate k to align with the dataset’s inherent structure and noise levels, demonstrating how KNN’s simplicity hinges critically on this choice.

# Splitting the 80% for training and 20% for testing

```python
import numpy as np

def split_data(x, y, seed, split_ratio=0.2):
    np.random.seed(seed)
    indices = np.random.permutation(len(y))
    split_index = int(split_ratio * len(y))
    
    # Split the data
    train_indices = indices[split_index:]
    test_indices = indices[:split_index]

    x_train_data = x[train_indices]
    x_test_data = x[test_indices]

    y_train_data = y[train_indices]
    y_test_data = y[test_indices]
    
    return x_train_data, x_test_data, y_train_data, y_test_data

# Example usage
x, y = open_200_dataset()
seeds = range(10)
k_values = [3, 5, 7, 9, 11, 13, 15]
mse_by_k = {k: [] for k in k_values}

for seed in seeds:
    x_train_data, x_test_data, y_train_data, y_test_data = split_data(x, y, seed)

    for k in k_values:
        new_regression_model = KNNRegressionModel(k)
        new_regression_model.fit(x_train_data, y_train_data)

        predicted_results = new_regression_model.predict(x_test_data)

        mse = new_regression_model.evaluate(y_test_data, predicted_results)

        mse_by_k[k].append(mse)

averages = [np.mean(mse_by_k[i]) for i in k_values]

print(averages)
```

# Implementing the FastKNN Classification logic

```python
from scipy.spatial import KDTree
from MachineLearningModel import MachineLearningModel
from collections import Counter
import numpy as np 
       
class FastKNNClassificationModel(MachineLearningModel):
    def __init__(self, k):
        self.k = k
        self.X_train = None
        self.y_train = None
        self.tree = None
        self.mean = None 
        self.std = None   

    def fit(self, X, y): 
        self.X_train = X
        self.y_train = y
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
        self.std = np.where(self.std == 0, 1.0, self.std)  
        
        X_norm = (X - self.mean) / self.std
        self.tree = KDTree(X_norm)  

    def predict(self, X):
        X_norm = (X - self.mean) / self.std
        _, indices = self.tree.query(X_norm, k=self.k)    
        
        predictions = []
        for neighbors in indices:
            neighbors = np.array([neighbors]) if np.isscalar(neighbors) else neighbors
            most_common = Counter(self.y_train[neighbors]).most_common(1)[0][0]
            predictions.append(most_common)
        
        return np.array(predictions)

    def evaluate(self, y_true, y_predicted):
        y_true = np.array(y_true)
        y_predicted = np.array(y_predicted)
        return np.sum(y_true == y_predicted)
```

# The main logic

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.io import loadmat  


mnist = loadmat("mnist-original.mat")  
mnist_data = mnist["data"].T 
mnist_label = mnist["label"][0]


X = mnist_data / 255.0
y = mnist_label.astype(int)


def train_test_split(X, y, test_size=1000, train_size=10000, random_state=None):
    if random_state:
        np.random.seed(random_state)
    indices = np.random.permutation(len(X))
    train_idx = indices[:train_size]
    test_idx = indices[train_size:train_size+test_size]
    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=1000, train_size=10000, random_state=42
)


def evaluate_knn(k_values, X_train, y_train, X_test, y_test):
    accuracies = []
    times = []
    
    for k in k_values:
        print(f"Evaluating k={k}...")
        model = FastKNNClassificationModel(k)
        
        
        start_time = time.time()
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
        elapsed = time.time() - start_time
        
        accuracy = model.evaluate(y_test, predictions)
        accuracies.append(accuracy)
        times.append(elapsed)
        print(f"k={k}: Correct Results = {accuracy:.4f}, Time = {elapsed:.2f}s")
    
    return accuracies, times


k_values = [1, 3, 5, 7, 9]
accuracies, times = evaluate_knn(k_values, X_train, y_train, X_test, y_test)


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_values, accuracies, 'bo-')
plt.xlabel('k value')
plt.ylabel('Accuracy')
plt.title('Accuracy vs k value')
plt.xticks(k_values)

plt.subplot(1, 2, 2)
plt.plot(k_values, times, 'ro-')
plt.xlabel('k value')
plt.ylabel('Time (seconds)')
plt.title('Prediction Time vs k value')
plt.xticks(k_values)

plt.tight_layout()
plt.show()

# Find best k
best_k = k_values[np.argmax(accuracies)]
print(f"\nBest k value: {best_k} with accuracy {max(accuracies):.4f}")

# Visualize some predictions
model = FastKNNClassificationModel(best_k)
model.fit(X_train, y_train)
sample_indices = np.random.choice(len(X_test), 5, replace=False)

plt.figure(figsize=(15, 3))
for i, idx in enumerate(sample_indices):
    plt.subplot(1, 5, i+1)
    plt.imshow(X_test[idx].reshape(28, 28), cmap='gray')
    pred = model.predict(X_test[idx:idx+1])[0]
    true = y_test[idx]
    plt.title(f"Pred: {pred}\nTrue: {true}")
    plt.axis('off')
plt.show()
```
